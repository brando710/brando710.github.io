```python
x = int(input('Enter your Score:'))
if x>=93:
  grade = 'A'
elif 92>=x>=90:
  grade = 'A-'
elif 89>=x>=87:
  grade = 'B+'
elif 86>=x>=83:
  grade = 'B'
elif 82>=x>=80:
  grade = 'B-'
elif 79>=x>=77:
  grade = 'C+'
elif 76>=x>=73:
  grade = 'C'
elif 72>=x>=70:
  grade = 'C-'
elif 69>=x>=67:
  grade = 'D+'
elif 66>=x>=63:
  grade = 'D'
elif 62>=x>=60:
  grade = 'D-'
elif 59>=x:
  grade = 'F'
print('Your grade is a', grade)
```
